## Import $/$ prep

````python
from _tmp.mnist import iteration
from _tmp.gapvae import prepare_stuff, validate
from _tmp.plotters import plot_latent_space, plot_latent_scatter

from torchvision.utils import make_grid
````

````python
batch_size = 200
epochs = 300

kws = dict(
    n_latents=None,
    beta=None,
    device_idx=1,
    model='pvae',
    activation='exp',
    weight_norm=True,
    max_grad_norm=300,
    batch_size=batch_size,
    epochs=epochs,
    lr=1e-3,
)
````

````python
%%time

results = collections.defaultdict(dict)
n_latents = [10, 50, 100, 1000, 5000]
betas = [0.7, 1.0]  # [0.1, 0.4]
looper = itertools.product(n_latents, betas)
# looper = tqdm(looper, ncols=130, position=0)
for n, b in looper:
    key = (n, b)
    # looper.set_description(f'n = {n} —— beta = {b:0.1f}')

    kws['n_latents'], kws['beta'] = n, b
    pvae, optim, optim_schedule, trn, vld, device, kws_iter = prepare_stuff(**kws)

    pbar = tqdm(range(epochs), ncols=138, position=0, leave=True)
    for ep in pbar:
        grad, tot, recon, kl = iteration(
            ep, pvae, trn, optim, optim_schedule, device, **kws_iter)

        msg = ',  '.join([
            f"latents, beta = ({n}, {b:0.1f})",
            f"grad = {grad:0.1f}",
            f"loss = (kl: {kl:0.1f}, recon: {recon:0.1f}  ——  tot: {tot:0.2f})  ",
        ])
        pbar.set_description(msg)

    
    emb, labels, loss_vld = validate(pvae, vld, device)
    results['loss_vld'][key] = loss_vld.copy()

    
    spks, r_q, r_p, r_qp = [], [], [], []
    for x, _ in iter(vld):
        x = x.to(device)
        with torch.inference_mode():
            out = pvae(x)
        spks.append(to_np(out['spks']))
        r_q.append(to_np(out['r_q']))
        r_p.append(to_np(out['r_p']))
        r_qp.append(to_np(out['r_q'] * out['r_p']))
    
    spks, r_q, r_p, r_qp = cat_map([spks, r_q, r_p, r_qp])
    results['spks'][key] = spks.copy()
    results['r_q'][key] = r_q.copy()
    results['r_p'][key] = r_p.copy()
    results['r_qp'][key] = r_qp.copy()

    
    fig, axes = create_figure(2, 2, (8, 4.5), layout='constrained')
    nbins = 500
    _kws = dict(
        legend=False, 
        stat='percent',
        bins=np.linspace(0, nbins, nbins+1) - 0.5,
        color='k',
    )
    sns.histplot(spks.ravel(),  ax=axes[0, 0], **_kws)
    _kws = dict(
        kde=True,
        legend=False, 
        stat='percent',
        line_kws={'lw': 2},
        alpha=0.2,
    )
    sns.histplot(r_q.ravel(), ax=axes[1, 0], **_kws)
    sns.histplot(np.log(r_qp.ravel()),  ax=axes[0, 1], **_kws)
    sns.histplot(np.log(r_p.ravel()), ax=axes[1, 1], **_kws)
    
    axes[0, 0].set_title('Spike counts')
    axes[0, 1].set_title(r'$r \delta r$' + ' (posterior)')
    
    axes[1, 0].set_title(r'$\delta r$' + ' (encoder output)')
    axes[1, 1].set_title(r'$r$' + ' (from prior)')

    for ax in axes[:, 1]:
        ax.set_ylabel('')
    
    add_grid(axes)
    
    plt.show()

    
    x_samples = pvae.sample(50)
    grid_images = make_grid(x_samples, nrow=10, normalize=True, value_range=(0, 1))

    x2p = to_np(grid_images)
    x2p = np.transpose(x2p, (1, 2, 0))
    
    fig, ax = create_figure(1, 1, (10, 3))
    ax.imshow(x2p)
    remove_ticks(ax)
    plt.show()

    
    x, _ = next(iter(vld))
    x = x.to(device)
    with torch.inference_mode():
        out = pvae(x)
    
    num = 32
    grid_images = make_grid(
        torch.cat([x[:num], out['x_recon'][:num]]),
        nrow=num,
        normalize=True,
        value_range=(0, 1),
    )
    
    x2p = to_np(grid_images)
    x2p = np.transpose(x2p, (1, 2, 0))
    
    fig, ax = create_figure(1, 1, (13, 5))
    ax.imshow(x2p)
    remove_ticks(ax)
    plt.show()
````
