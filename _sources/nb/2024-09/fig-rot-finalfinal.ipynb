````python
df = pd.read_pickle(pjoin(tmp_dir, 'df_rot.df'))
````

````python
rename_marino_models = {
    'ia-VAE (single-level)': 'ia-VAE (s)',
    'ia-VAE (hierarchical)': 'ia-VAE (h)',
}
````

````python
pal_model = get_palette_models()

for name, new_name in rename_marino_models.items():
    df.loc[df['model'] == name, 'model'] = new_name
    pal_model[new_name] = pal_model.pop(name)
````

````python
fig, axes = create_figure(2, 1, (10, 7), sharex='col')

_df = df.loc[
    (df['model'] != 'aiP-VAE')
]
for i, y_name in enumerate(['mse', 'accuracy']):
    ax = axes[i]
    sns.barplot(
        data=_df,
        x='theta',
        y=y_name,
        hue='model',
        legend=i==0,
        palette=pal_model,
        ax=ax,
    )
    ax.grid(axis='y')
# axes[0].set_yscale('log')
axes[1].set_ylim((0.57, 1.0))
axes[1].set_xlabel(r'$\theta$', fontsize=15)

sns.move_legend(axes[0], 'upper right', fontsize=12)

# axes[0].set_ylabel('Reconstruction error (MSE)', fontsize=11)
# axes[1].set_ylabel('Classification accuracy', fontsize=12)

axes[0].set_ylabel(' ', fontsize=11)
axes[1].set_ylabel(' ', fontsize=12)

fig.savefig(pjoin(fig_dir, 'rot-bar.pdf'), **_KWS_FIG)
fig.savefig(pjoin(fig_dir, 'rot-bar.png'), **_KWS_FIG)

plt.show()
````
