## Save the fig

````python
df = pd.read_pickle(pjoin(tmp_dir, 'df_ood.df'))
````

````python
rename_marino_models = {
    'ia-VAE (single-level)': 'ia-VAE (s)',
    'ia-VAE (hierarchical)': 'ia-VAE (h)',
}
````

````python
pal_model = get_palette_models()

for name, new_name in rename_marino_models.items():
    df.loc[df['model'] == name, 'model'] = new_name
    pal_model[new_name] = pal_model.pop(name)
````

````python
fig, axes = create_figure(1, 2, (5.2, 3), width_ratios=[2.0, 1])

_df = df.loc[
    (df['model'] != 'aiP-VAE') &
    (df['ds_src'] == 'MNIST') &
    (df['ds_tgt'].isin(['EMNIST', 'Omniglot']))
]
sns.barplot(
    data=_df,
    x='ds_tgt',
    y='mse',
    hue='model',
    palette=pal_model,
    legend=False,
    ax=axes[0],
)

_df = df.loc[
    (df['model'] != 'aiP-VAE') &
    (df['ds_src'] == 'MNIST') &
    (df['model'] != 'aiP-VAE') &
    (df['ds_tgt'] == 'EMNIST')
]
sns.barplot(
    data=_df,
    x='ds_tgt',
    y='accuracy',
    hue='model',
    palette=pal_model,
    legend=True,
    ax=axes[1],
)
for ax in axes.flat:
    ax.set(xlabel='')
    ax.grid(axis='y')

sns.move_legend(axes[1], 'best', fontsize=10)
move_legend(axes[1], (1, 1.03))

# axes[0].set_ylabel('Reconstruction error (MSE)', fontsize=11)
# axes[1].set_ylabel('Classification accuracy', fontsize=12)
axes[0].set_ylabel(' ', fontsize=11)
axes[1].set_ylabel(' ', fontsize=12)

fig.savefig(pjoin(fig_dir, 'ood-bar.pdf'), **_KWS_FIG)
fig.savefig(pjoin(fig_dir, 'ood-bar.png'), **_KWS_FIG)

plt.show()
````
