````python
device_idx = 1
device = f'cuda:{device_idx}'

from analysis.final import sort_fits, analyze_fits
````

````python
root = 'Dropbox/chkpts/PoissonVAE_may22_submitted'
fits, fits_st, fits_etc = sort_fits(root=root)
len(fits), len(fits_st), len(fits_etc)
````

````python
def accept_fn(name):
    archi = name.split('-')[1]
    accept = (
        'MNIST' in name and
        archi == '<conv+b|conv+b>'
    )
    return accept


fits = sorted(filter(accept_fn, fits))
fits_etc = sorted(filter(accept_fn, fits_etc))

print(f"# fits: {len(fits)},  # fits_etc: {len(fits_etc)}")
````

````python
%%time

kws = dict(device=device, analysis_mode='clf', clf_type='logreg', root=root)
df = pd.concat([analyze_fits(fits, **kws), analyze_fits(fits_etc, **kws)])

df = df.sort_values(by=['type', 'seed', 'n_dims'])
df = df.reset_index(drop=True)

save_obj(
    obj=df,
    file_name='df_clf',
    save_dir=tmp_dir,
    verbose=True,
    mode='df',
)
````
